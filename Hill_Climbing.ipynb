```python
import random

def objective_function(x):
    """
    This is the 'hill' we are climbing.
    We want to find the maximum value of: f(x) = -(x - 5)^2 + 20
    The peak is at x = 5.
    """
    return -(x - 5)**2 + 20

def hill_climbing():
    # 1. Start at a random initial position between -10 and 10
    current_x = random.uniform(-10, 10)
    current_value = objective_function(current_x)

    step_size = 0.1  # How far we look to the left and right
    max_iterations = 1000

    print(f"Starting Position: x = {current_x:.2f}, Value = {current_value:.2f}")

    for i in range(max_iterations):
        # 2. Look at neighbors
        next_x_right = current_x + step_size
        next_x_left = current_x - step_size

        val_right = objective_function(next_x_right)
        val_left = objective_function(next_x_left)

        # 3. Move to the best neighbor if it improves our height
        if val_right > current_value and val_right >= val_left:
            current_x = next_x_right
            current_value = val_right
        elif val_left > current_value:
            current_x = next_x_left
            current_value = val_left
        else:
            # 4. If no neighbor is higher, we have reached a peak
            print(f"Reached peak at iteration {i}")
            break

    return current_x, current_value

# Run the search
best_x, best_val = hill_climbing()

print("---------------------------------")
print(f"Optimal Solution: x = {best_x:.4f}")
print(f"Maximum Height (Value): {best_val:.4f}")
```
