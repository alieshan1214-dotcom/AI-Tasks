```python
import heapq

# Define the graph as an adjacency list where each key is a node and the value is a list of tuples (neighbor, edge_weight)
# Note: For Uniform Cost Search, we do not need heuristics, only edge weights.
graph = {
    # Node S connects to A with weight 2 and B with weight 5
    'S': [('A', 2), ('B', 5)],
    # Node A connects to C with weight 2 and D with weight 4
    'A': [('C', 2), ('D', 4)],
    # Node B connects to D with weight 5 and E with weight 1
    'B': [('D', 5), ('E', 1)],
    # Node C connects to G with weight 3
    'C': [('G', 3)],
    # Node D connects to G with weight 2
    'D': [('G', 2)],
    # Node E connects to G with weight 6
    'E': [('G', 6)],
    # Node G has no outgoing edges as it is the goal
    'G': []
}

# Define the Uniform Cost Search (UCS) function
def uniform_cost_search(graph, start, goal):
    # Initialize the open list (priority queue)
    # The tuple format used is: (cumulative_cost, current_node, path_taken_so_far)
    open_list = []

    # Push the start node into the open list with an initial cost of 0
    heapq.heappush(open_list, (0, start, [start]))

    # Initialize a dictionary to keep track of the minimum cost to reach each node
    # This acts as our visited set and cost tracker combined
    costs = {start: 0}

    # Loop until the priority queue is empty
    while open_list:
        # Pop the node with the lowest cumulative cost
        current_cost, current_node, current_path = heapq.heappop(open_list)

        # Check if we have reached the goal node
        if current_node == goal:
            # Return the optimal path and its total cost
            return current_path, current_cost

        # Explore neighbors
        for neighbor, weight in graph[current_node]:
            # Calculate the total cost to reach the neighbor
            new_cost = current_cost + weight

            # If we found a new node or a cheaper path to an already visited node
            if neighbor not in costs or new_cost < costs[neighbor]:
                # Update the known lowest cost to reach this neighbor
                costs[neighbor] = new_cost

                # Append the neighbor to the current path
                new_path = current_path + [neighbor]

                # Push the neighbor into the priority queue to be explored later
                heapq.heappush(open_list, (new_cost, neighbor, new_path))

    # If the queue is empty and the goal hasn't been reached
    return None, float('inf')

if __name__ == "__main__":
    start_node = 'S'
    goal_node = 'G'

    path, cost = uniform_cost_search(graph, start_node, goal_node)

    if path:
        print(f"Optimal Path found: {' -> '.join(path)}")
        print(f"Total Cost: {cost}")
    else:
        print("Goal is not reachable from the start node.")
```
